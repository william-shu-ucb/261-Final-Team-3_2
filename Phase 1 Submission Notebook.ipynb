# Abstract



## Abstract

Flight delays represent a major operational challenge for airlines and airports, leading to increased operational costs, inefficient resource allocation, and reduced passenger satisfaction. In this project, we aim to develop a scalable machine learning pipeline to predict airline flight delays using the On-Time Performance and Weather (OTPW) dataset. This dataset integrates airline operational data, airport metadata, and meteorological observations, enabling analysis of how environmental conditions and operational factors influence flight delays.

Exploratory data analysis reveals several important characteristics of the dataset. The distribution of departure delays is highly **right-skewed**, with most flights experiencing little or no delay while a smaller subset experiences significant disruptions. Additionally, analysis of the weather station–airport mapping shows that most airports rely on nearby weather stations for observations, although some airports depend on stations located farther away. These findings suggest that **weather conditions, temporal patterns, and operational factors** are likely important drivers of delay behavior.

To address this problem, we will build a distributed machine learning pipeline using Apache Spark to efficiently process large-scale aviation datasets. The primary modeling task will be formulated as a regression problem, predicting the continuous delay variable (`DEP_DELAY`) in minutes. Predicted delay values will then be converted into **operational delay categories (0–15, 15–45, and 45+ minutes)** to support decision-making by airlines and airport operators. We will evaluate several machine learning models, including **linear regression, logistic regression, tree-based ensemble models (Random Forest, Gradient Boosted Trees, and LightGBM), and neural networks**, and compare their performance using appropriate regression and classification metrics. The goal is to identify key drivers of flight delays and develop predictive models that can support **data-driven operational planning in the aviation industry**.

<br>

### Problem Statement

Flight delays create cascading disruptions across the aviation system, affecting airlines, airports, and passengers. Delays increase operational costs, reduce aircraft and crew utilization efficiency, and negatively impact passenger satisfaction. Being able to accurately predict flight delays before departure would allow airlines and airport operators to proactively manage resources, communicate with passengers, and adjust schedules to mitigate downstream disruptions.

In this project, we aim to develop a machine learning pipeline capable of **predicting the expected departure delay of a flight in minutes**. The prediction must be generated **at least two hours prior to the scheduled departure time**, meaning that the model can only use information available before that point. This constraint ensures that the prediction system reflects a realistic operational setting for airline decision-making.

To address this problem, we use the **On-Time Performance and Weather (OTPW) dataset**, which integrates airline operational data with meteorological observations. The dataset enables analysis of how factors such as **weather conditions, airport characteristics, airline operations, and temporal patterns** influence flight delay behavior.

The primary modeling task is formulated as a **regression problem**, where the model predicts the continuous delay variable `DEP_DELAY` (measured in minutes). For operational interpretation, the predicted delay values will also be mapped into **three delay categories**:  
- **0–15 minutes:** No delay  
- **15–45 minutes:** Moderate delay  
- **45+ minutes:** Major delay  

Because aviation datasets are large and complex, we will build a scalable data processing and modeling pipeline using **Apache Spark**, enabling efficient distributed processing of large-scale flight and weather data.

<br>

### Stakeholders

Several stakeholders could benefit from accurate flight delay prediction models:

#### Airlines
Airlines can use delay predictions to improve operational planning and reduce disruption costs. Accurate predictions allow airlines to better manage aircraft rotations, crew scheduling, and passenger communication.

#### Airport Authorities
Airport operators can use delay forecasts to improve gate allocation, runway scheduling, and ground service coordination, helping reduce congestion during peak operational periods.

#### Air Traffic Control and Aviation Regulators
Organizations responsible for managing airspace and ensuring safety can benefit from predictive insights that help anticipate congestion and coordinate traffic flow more effectively.

#### Passengers
Passengers benefit from earlier delay notifications, allowing them to adjust travel plans, rebook flights, or make alternative arrangements.

Overall, predictive delay models can contribute to **improved operational efficiency and a better passenger experience across the aviation ecosystem.**

The **Stations dataset (`df_stations`)** and the **Airport Codes dataset (`df_airport_codes`)** are used together to link flight operations with nearby weather observations. These datasets provide the foundation for integrating meteorological conditions into the flight delay prediction problem.

The **Stations dataset** maps weather stations to nearby airports and records the distance between them. Each row represents a relationship between a weather station and an airport, allowing weather observations from the station to be used as a proxy for airport weather conditions. The dataset contains more than 5 million station–airport relationships, providing broad geographic coverage.

Analysis of the `distance_to_neighbor` variable shows that the average station-to-airport distance is approximately 1,343 units, with considerable variation across airports. Distances range from 0 (stations located directly at airports) to more than 6,400 units, indicating that some airports rely on weather stations that are relatively far away. This variability suggests that weather measurements may differ in representativeness depending on station proximity, making station distance a potentially useful feature or filtering criterion when constructing weather-based features.

The **Airport Codes dataset** provides metadata describing airports around the world. Key attributes include airport identifiers (`ident`), airport type, geographic region, municipality, elevation, and geographic coordinates. While the dataset contains global airport metadata, it is heavily skewed toward small airports and heliports, with only about **600 airports classified as large airports. These large airports represent the most relevant subset for modeling scheduled commercial airline operations.

# Machine Algorithms and Metrics

#### 1. Machine Learning Algorithms

##### Problem Formulation

A key challenge in the flight delay prediction problem is that predictions must be generated **at least two hours before a flight’s scheduled departure**, meaning that the model can only use information available prior to that time. This constraint introduces temporal dependencies because flight delays often propagate through the aviation system due to weather conditions, airport congestion, aircraft rotations, and operational disruptions.

In this project, we treat the primary modeling task as a **regression problem**, where the goal is to predict the continuous delay variable `DEP_DELAY` (measured in minutes). Predicting the exact delay duration allows the model to capture the continuous nature of delay behavior and provides more detailed insight into operational conditions affecting flights. Additionally, flight delay data are typically **right-skewed**, with many flights experiencing little or no delay and a smaller number of flights experiencing large disruptions. This distribution motivates the use of models capable of handling nonlinear relationships and extreme values.

For operational planning and communication purposes, the predicted delay values will then be converted into **three delay categories**:

- **0–15 minutes:** No delay  
- **15–45 minutes:** Moderate delay  
- **45+ minutes:** Major delay  

These bins provide meaningful delay ranges that are easier for airlines, airport operators, and passengers to interpret and use for planning. This approach allows us to maintain the fidelity of a continuous prediction while translating results into **actionable delay categories**.

Temporal information will be incorporated through features such as **departure time, day-of-week patterns, airport congestion indicators, and historical delay propagation patterns**.

##### Baseline Models

We will begin with simple models to establish performance benchmarks for the regression formulation of the problem.

**Linear Regression**

- Implementation: Spark ML or scikit-learn  
- Objective: Predict continuous delay minutes (`DEP_DELAY`)  
- Loss Function: Mean Squared Error (MSE)

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

Linear regression provides a simple and interpretable benchmark that allows us to compare the effectiveness of more complex nonlinear models.

##### Supporting Classification Baseline

To complement the regression analysis, we will also evaluate a classification baseline using the delay categories derived from `DEP_DELAY`.

**Logistic Regression (Multinomial)**

- Implementation: Spark ML or scikit-learn  
- Objective: Predict delay categories derived from the operational delay bins  
- Loss Function: Log Loss

$$
L = -\frac{1}{n} \sum_{i=1}^{n} \left[y_i \log(\hat{y}_i) + (1-y_i)\log(1-\hat{y}_i)\right]
$$

Logistic regression provides a simple and interpretable baseline for delay classification and allows us to evaluate whether more complex models significantly improve predictive performance.

##### Tree-Based Ensemble Models

Tree-based ensemble models are expected to perform well because they can capture **nonlinear relationships between weather conditions, airport characteristics, and operational factors**.

Airline delay data involve complex interactions between variables such as weather conditions, airport congestion, carrier operations, and time-of-day patterns. Tree-based models can automatically capture these nonlinear relationships and feature interactions without requiring extensive manual feature engineering, which often leads to stronger predictive performance on structured tabular datasets.

The following models will be evaluated:

  - Bagging ensemble of decision trees
  - Helps reduce variance and prevent overfitting

- **Gradient Boosted Trees**
  - Sequential boosting approach that reduces residual errors
  - Often provides strong predictive performance for tabular data

- **LightGBM**
  - Histogram-based gradient boosting framework
  - Efficient training on large datasets and well-suited for distributed environments

### 2. Successful Evaluation Metrics

Because flight delays represent a relatively **rare event**, class imbalance is an important consideration when evaluating model performance. Therefore, we will use multiple evaluation metrics to assess how well the models identify delayed flights while minimizing false predictions.

##### Classification Metrics

Since the primary task is to predict whether a flight will experience a significant delay (≥15 minutes), we will focus on the following classification metrics:

**Accuracy**

$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$

Accuracy measures the proportion of correct predictions among all predictions.

**Precision**

$$
Precision = \frac{TP}{TP + FP}
$$

Precision measures the proportion of predicted delayed flights that are actually delayed. High precision reduces false alarms.

**Recall**

$$
Recall = \frac{TP}{TP + FN}
$$

Recall measures the proportion of actual delayed flights that are correctly identified. This metric is particularly important because failing to identify delays could lead to operational disruptions.

**F1 Score**

$$
F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}
$$

The F1 score balances precision and recall, providing a single measure of overall classification performance.

**ROC-AUC**

The Area Under the Receiver Operating Characteristic Curve (ROC-AUC) measures the model’s ability to distinguish between delayed and on-time flights across different classification thresholds.

### Regression Metrics

In addition to classification, we will experiment with regression models that predict the continuous delay value (`DEP_DELAY`). These predictions may later be converted into delay categories for operational interpretation.

**Mean Squared Error (MSE)**

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

MSE measures the average squared difference between predicted and actual delay values.

**Mean Absolute Error (MAE)**

$$
MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

MAE measures the average absolute prediction error and is less sensitive to extreme outliers than MSE.

### 3. Modeling Plan

Our modeling workflow combines both **regression and classification approaches** to capture the continuous nature of delay times while producing operationally meaningful delay categories for decision-making.

The modeling process will follow these steps:

1. **Define the operational delay targets.**  
   We define a **binary delay indicator using the 15-minute threshold** (delayed vs. on-time) while also preserving the continuous target variable `DEP_DELAY` for regression modeling.

2. **Train baseline models.**  
   We begin with interpretable baseline models such as **Linear Regression** for delay minutes and **Logistic Regression** for delay classification to establish performance benchmarks.

3. **Train more advanced machine learning models.**  
   We then train ensemble-based models including **Random Forest, Gradient Boosted Trees, and LightGBM** to capture nonlinear relationships and complex interactions among features.

4. **Evaluate regression performance.**  
   For models predicting `DEP_DELAY`, we evaluate performance using metrics such as **Mean Squared Error (MSE)** and **Mean Absolute Error (MAE)**.

5. **Convert regression predictions into operational delay categories.**  
   Predicted delay minutes are mapped into operational delay categories to support downstream classification evaluation and operational interpretation.

6. **Evaluate classification performance.**  
   Models are evaluated using metrics including **precision, recall, F1 score, and ROC-AUC**, with particular emphasis on **recall and F1 score** due to class imbalance in delay events.

7. **Compare models and analyze feature importance.**  
   We select the best-performing model based on overall predictive performance and analyze **feature importance** to identify key drivers of flight delays.

Overall, the goal is to develop a model that produces **reliable early predictions of flight delays while minimizing false alarms and missed delay events**, enabling more effective operational planning for airlines and airports.

# Machine Learning Pipeline

Our pipeline transforms raw flight and weather data into delay predictions through the following steps:

1. **Data Ingestion & Checkpointing.**  
   Raw data is ingested in Parquet format using Apache Spark. Key transformation results are checkpointed to disk after expensive operations (e.g., null analysis, type casting) to avoid recomputation.

2. **Data Cleaning & Preprocessing.**  
   Cancelled flights and columns with >50% missing values are removed. Regex-based cleaning and type casting are applied to string-encoded numeric columns. All **22 leakage columns** that would not be available two hours before departure are excluded.

3. **Feature Engineering.**  
   Features are constructed across four categories: **temporal** (departure hour, day of week, month), **weather** (hourly temperature, wind speed, visibility, precipitation, humidity, pressure), **operational** (carrier, origin/destination airport, scheduled elapsed time, distance), and **historical** (rolling average delay by carrier and airport).

4. **Train / Validation / Test Split.**  
   Data is split **chronologically** to prevent data leakage — training on earlier periods and validating/testing on later ones. Normalization statistics (mean, std) are computed on the training set only and applied to validation and test sets.

5. **Model Training.**  
   We train and compare baseline models (**Linear Regression**, **Logistic Regression**), ensemble methods (**Random Forest**, **Gradient Boosted Trees**, **LightGBM**), and a **Multilayer Perceptron (MLP)** neural network.

6. **Model Evaluation.**  
   The target variable is **`DEP_DEL15`** — a binary indicator of whether a flight is delayed by 15 minutes or more. Models are evaluated using **precision, recall, F1 score, and ROC-AUC**, with emphasis on recall due to the class imbalance between on-time (**80%**) and delayed (**20%**) flights.

7. **Feature Importance Analysis.**  
   We analyze feature importance from the best-performing model to identify the key drivers of flight delays and refine the feature set for subsequent phases.

Overall, the goal is to develop a pipeline that produces **reliable early predictions of whether a flight will be delayed by 15 or more minutes**, enabling more effective operational planning for airlines and airports.
