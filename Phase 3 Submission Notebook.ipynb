# Data and Feature Engineering

### Data Lineage
Our analysis uses two datasets for our modeling pipeline. The first is the the Ontime Performance of Flights and Weather dataset (OTPW). It is a pre-joined dataset constructed by integrating three primary raw data sources: historical commercial flight records from U.S. DOT TranStats, hourly meteorological observations from NOAA's ISD, and geographic airport mapping data to link each flight to its nearest weather reporting station.

The second dataset is a weather dataset, including hourly meteorological observations from NOAA's ISD. This is the same dataset as the raw source joined into the OTPW dataset. However, we noticed leakage concerns with the first join, so we decided to incorperate more data to from this weather dataset to fix these issues. This included pulling weather information for each flight that was more than two hours before departure time. More information about this will be discussed in the leakage section below.

### Key Transformations

To transform and join the OTPW and Weather datasets, we first performed cleaning. On the OTPW dataset, we deduplicated flights, removed cancelled flights, and only retained flights with hourly weather information. We also created timestamp columns from flight departure times, including taking the earliest of scheduled/actual departure time and shifting the time back by two hours. Finally we extracted the hour from the shifted timestamps to use during our join.

On the weather dataset, we filtered for only hourly weather information, created timestamp columns from date infomation, and extracted the hour from the timestamp.

After cleaning, we were able to left join OTPW with weather, on the created hourly timestamp variable (which was shifted by two hours on the OPTW dataset) and weather station ID. This join was able to fix some of our leakage concerns regarding weather in the dataset.

**DISCUSSION ABOUT PAGERANK FEATURES**

To capture the structural importance of each airport, we created a group of graph-based PageRank features that quantify how central or influential an airport is based on the pattern of flights connecting origins and destinations. The rationale is that delays propagate more readily through highly connected hubs, so incorporating network centrality should help the model detect airports that are more prone to cascading disruptions.

To implement PageRank, we constructed a directed graph for each of data where each airport is a vertex and each flight route is an edge. The flights were aggregated by route to form a list of edges before we built the corresponding vertices list. From this graph, we computed unweighted and weighted PageRank scores. Unweighted PageRank treats each route equally where airports with many inbound or outbound connections receive higher scores regardless of flight volume. Weighted PageRank duplicates edges according to the number of flights on each route, allowing edges with more traffic to exert more influence on the centrality calcuation. This distinction allows the model to learn both structural connectivity and traffic-adjusted influence.

After computing PageRank for each year, we joined the resulting scores back to our dataset twice: once on the origin airport and once on the destination airport. This produced four final features: `origin_pagerank_unweighted`, `dest_pagerank_unweighted`, `origin_pagerank_weighted`, and `dest_pagerank_weighted`. Each feature provides a different perspective on how central an airport is within the network at the time of departure. For example, a flight departing from a high-PageRank origin airport may be more susceptible to upstream congestion, while a flight arriving at a high-PageRank destination may face increased risk of downstream bottlenecks.
