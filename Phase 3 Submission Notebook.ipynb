# Project Abstract

**The final form of the abstract! It should have everything covered in previous phases, plus the new experiments and the final model selected, as well as your final results (report the number!)**

Flight delays create cascading operational disruptions for airlines, impacting crew scheduling, aircraft rotations, gate availability, and passenger connections. These challenges increase operating costs and reduce customer satisfaction. In this project, we focus on the Operations Control Center (OCC) Team of Airline XYZ as the primary stakeholder. We formulate the task as a binary classification problem, predicting whether a U.S. domestic flight will depart 15 minutes or more late, two hours before scheduled departure, using only information available at that time. Early and accurate predictions enable proactive adjustments to staffing, gate assignments, aircraft flow, and passenger recovery plans, improving operational efficiency and reducing financial risk.

For modeling, we use an integrated dataset combining U.S. Department of Transportation flight performance records with National Oceanic and Atmospheric Administration (NOAA) hourly weather observations, along with airport, temporal, congestion, graph-based, and newly engineered trend features. Given the higher operational cost of missed delays compared with false alarms, we prioritize F2-score as the primary evaluation metric and Recall as a secondary metric. Logistic Regression was selected as the initial baseline due to its interpretability and role as a standard benchmark, achieving F2 = 0.475 and Recall = 0.72 on the blind test set. We then evaluated advanced models including Random Forest, Gradient Boosted Trees, and LSTM neural networks. Incorporating weather trend and tail-number propagation features materially improved model performance, increasing Random Forest from 0.4903 to 0.6006, Gradient Boosted Trees from 0.4974 to 0.6198, and LSTM from 0.5407 to 0.6143 on the 2019 blind test set. Our final selected model is Gradient Boosted Trees, which achieved the strongest overall performance with F2 = 0.6198 while maintaining strong recall, successfully identifying the majority of delayed flights. Key predictive features include airport network importance (PageRank), short-term airport congestion (2–4 hour prior delay counts), two-hour lagged weather conditions, and newly engineered weather trend signals.

Across this project, we developed a scalable machine learning pipeline, implemented imbalance-aware techniques, engineered operationally relevant features, and validated performance using time-based cross-validation to prevent data leakage. Our final results confirm that predicting delays two hours in advance is both feasible and operationally meaningful. More importantly, the findings show that feature engineering contributed more to performance gains than model complexity alone. Future work will focus on expanded aircraft propagation features, destination congestion signals, and advanced ensemble approaches to further improve predictive performance and support production-ready OCC decision workflows.



# Data and Feature Engineering

### Data Lineage
Our analysis uses two datasets for our modeling pipeline. The first is the the Ontime Performance of Flights and Weather dataset (OTPW). It is a pre-joined dataset constructed by integrating three primary raw data sources: historical commercial flight records from U.S. DOT TranStats, hourly meteorological observations from NOAA's ISD, and geographic airport mapping data to link each flight to its nearest weather reporting station.

The second dataset is a weather dataset, including hourly meteorological observations from NOAA's ISD. This is the same dataset as the raw source joined into the OTPW dataset. However, we noticed leakage concerns with the first join, so we decided to incorperate more data to from this weather dataset to fix these issues. This included pulling weather information for each flight that was more than two hours before departure time. More information about this will be discussed in the leakage section below.

### Key Transformations

To transform and join the OTPW and Weather datasets, we first performed cleaning. On the OTPW dataset, we deduplicated flights, removed cancelled flights, and only retained flights with hourly weather information. We also created timestamp columns from flight departure times, including taking the earliest of scheduled/actual departure time and shifting the time back by two hours. Finally we extracted the hour from the shifted timestamps to use during our join.

On the weather dataset, we filtered for only hourly weather information, created timestamp columns from date infomation, and extracted the hour from the timestamp.

After cleaning, we were able to left join OTPW with weather, on the created hourly timestamp variable (which was shifted by two hours on the OPTW dataset) and weather station ID. This join was able to fix some of our leakage concerns regarding weather in the dataset.

### PageRank Graph Features Description + EDA

| **Features**                   |  **Description**            | **Type**                 |   **Example**           |
|-|-|-|-|
| `origin_pagerank_weighted`          | Weighted PageRank score of the origin airport. Edges are weighted by the number of flights between airports. Higher values indicate airports that are highly influential hubs in terms of traffic volume.  | Double | 0.043 |
| `origin_pagerank_unweighted`        | Unweighted PageRank score of the origin airport. Every route counts equally regardless of traffic volume. Captures pure centrality in the flight network. | Double | 0.005 |
| `dest_pagerank_weighted`            | Weighted PageRank score of the destination airport. Reflects how central and heavily trafficked the destination is within the network. | Double | 0.086 |
| `dest_pagerank_unweighted`          | Unweighted PageRank score of the destination airport. Measuring its structural importance in the network independent of flight counts. | Double | 0.0034  |

To capture the structural importance of each airport, we created a group of graph-based PageRank features that quantify how central or influential an airport is based on the pattern of flights connecting origins and destinations. The rationale is that delays propagate more readily through highly connected hubs, so incorporating network centrality should help the model detect airports that are more prone to cascading disruptions.

To implement PageRank, we constructed a directed graph for each of data where each airport is a vertex and each flight route is an edge. The flights were aggregated by route to form a list of edges before we built the corresponding vertices list. From this graph, we computed unweighted and weighted PageRank scores. Unweighted PageRank treats each route equally where airports with many inbound or outbound connections receive higher scores regardless of flight volume. Weighted PageRank duplicates edges according to the number of flights on each route, allowing edges with more traffic to exert more influence on the centrality calcuation. This distinction allows the model to learn both structural connectivity and traffic-adjusted influence.

After computing PageRank for each year, we joined the resulting scores back to our dataset twice: once on the origin airport and once on the destination airport. This produced four final features: `origin_pagerank_unweighted`, `dest_pagerank_unweighted`, `origin_pagerank_weighted`, and `dest_pagerank_weighted`. Each feature provides a different perspective on how central an airport is within the network at the time of departure. For example, a flight departing from a high-PageRank origin airport may be more susceptible to upstream congestion, while a flight arriving at a high-PageRank destination may face increased risk of downstream bottlenecks.

# Leakage

**-- Describe how your pipeline does not suffer from any leakage problem and does not violate any cardinal sins of ML**


Data leakage occurs when a model learns from information that would not be available at prediction time. A common example is accidentally including blind test-set data in the training data, which inflates model performance and leads to much worse results than expected. To avoid this, we regorously checked and removed several sources of data leakage from our feature engineering and modeling pipelines.

One major leakage risk came from the weather data in our initial On-Time Flight Performance and Weather (OTPW) dataset, with many of those observations being recorded after a flight's scheduled departure time. Because our goal is to predict delays two hours before departure, we redefined each flight's observation timestamp as two hours before the earlier of scheduled or actual departure, truncated to the hour. We then re-sourced hourly NOAA weather data, aligned the timestamps to the nearest hour, and joined only the weather records that would have been available at that earlier time with matching station IDs to the OTPW table, ensuring the model never sees future weather conditions.

We also examined leakage risks from features that depend on learned statistics, such as standardization and PageRank. Because we use rolling cross-validation, each training fold is processed independently only using information from its own training year (means, standard deviations, etc.) before being applied to the validation year. The final 2015-2018 statistics are fit only on the training period and applied to the 2019 test set, ensuring no information flows from test to train. 

PageRank features follow the same principle. Each year's airport network is built only from flights in that year, and PageRank scores are computed separately before being joined back to the corresponding year's data. 

In practice, data leakage can never be fully eliminated in applied machine learning, especially with time-dependent features. Ideally, PageRank for year _t_ would be computed using edges from year _t-1_. Because 2014 data is unavailable, our 2015 flights use PageRank computed from 2015 edges. Similarly, our 2016-2018 flights also use the same-year edges. We recognize that this introduces slight temporal leakage, and acknowledge that future work could reduce this further by computing PageRank quarterly and using the previous quarter's graph.

Overall, our pipeline avoids the cardinal sins of machine learning: we never train on test data, never compute statistics on the full dataset, and never allow future information to influence past predictions. With the exception of the small temporal limitation leakage in our PageRank implementation, our feature engineering and modeling workflow is designed to be leakage-free and faithful to real-world prediction conditions.


# Modeling Pipelines 
 
We ingest the pre-processed, feature-engineered training dataset (2015–2018) and a held-out blind test set (2019). Raw features are transformed through a Spark ML preprocessing sub-pipeline with three variants tailored to each model family then passed to a classifier. Three sub-pipelines were constructed: one for Logistic Regression, one for tree-based models (RF and GBT), and one for MLP. The LSTM uses a separate PyTorch pipeline outside of Spark ML entirely. All models share the same feature families but use three distinct preprocessing pipelines tailored to each model family.

###Logistic Regression###
In our baseline, numeric features are standardized and log-transformed where needed. Categorical features are one-hot encoded and assembled with numerics into a single feature vector. To avoid multicollinearity, the airport congestion score is replaced with its log-transformed equivalent.

###Random Forest & Gradient Boosted Trees###
Same preprocessing as Logisitc Regression, except both the raw congestion count and the congestion percentage are retained, tree models are immune to multicollinearity and benefit from having both signals.

###MLP###
The origin airport is dropped from one-hot encoding since 300+ airports produce a very high-dimensional sparse vector that degrades network performance. Continuous PageRank scores represent airport importance instead. Two architectures are tested: one hidden layer and two hidden layers.

###LSTM###
Only numeric features are used — no one-hot encoding. Flights are grouped by aircraft and sorted chronologically, then assembled into sequences of 5 consecutive legs per aircraft to capture delay propagation across an aircraft's daily schedule. Sequences are zero-padded when fewer than 5 prior legs exist.

### Loss Functions

**Logistic Regression** — regularized log-loss:

$$\mathcal{L}(\mathbf{w}) = -\frac{1}{n} \sum_{i=1}^{n} \left[ y_i \log \hat{p}_i + (1-y_i)\log(1-\hat{p}_i) \right] + \lambda \left[ \alpha \|\mathbf{w}\|_1 + \frac{1-\alpha}{2}\|\mathbf{w}\|_2^2 \right]$$

where \\(\lambda\\) is regularization strength and \\(\alpha\\) controls the L1/L2 mix.

**GBT** — logistic loss across boosting rounds:

$$\mathcal{L} = \sum_{i=1}^{n} \log\left(1 + e^{-y_i F(\mathbf{x}_i)}\right)$$

where \\(F(\mathbf{x}_i)\\) is the ensemble prediction scaled by the learning rate. Complexity is controlled via tree depth and learning rate.

**LSTM** — binary cross-entropy with logits:

$$\mathcal{L} = -\frac{1}{n} \sum_{i=1}^{n} \left[ y_i \log \sigma(z_i) + (1-y_i)\log(1-\sigma(z_i)) \right]$$

where \\(z_i\\) is the raw output of the final fully-connected layer. Dropout acts as an implicit regularizer.

### Cross-Validation Consistency Across Folds

All Phase III experiments used **3 rolling year-based folds** on the 2015–2018 training window: Train 2015 → Validate 2016, Train 2016 → Validate 2017, and Train 2017 → Validate 2018. The 2019 blind test was held out completely. This temporal split is the right cross-validation design for a forecasting problem because each fold trains only on data that would have been available at the time of prediction; a random k-fold split would leak future information backward and inflate metrics. The same fold definitions are used for V1, V2, and V3, so any difference in CV performance reflects the feature set and not the validation protocol.

The champion V2 GBT shows the kind of consistency we want from a model that will be deployed: per-fold Test F2 stays inside a roughly 0.02 band, the train-vs-validation gap is similar across folds (no fold catastrophically overfits while others generalize), and the model rank ordering (GBT ≥ LSTM > MLP Deep > LR) is preserved fold-to-fold rather than flipping. The 2019 blind-test F2 = 0.62 lands in the same band as the 3-fold CV mean, confirming that the blind-test number is not a lucky single-year artifact.

The V1 → V2 transition was validated through CV before the blind test: across all four candidate models (LR, GBT, MLP Deep, LSTM), V2 raised mean CV Test F2 with the rank ordering preserved on every fold. That cross-fold consistency is what gave us confidence to commit V2 to the blind test rather than treat the improvement as noise. The reverse logic applied to V3 — its CV Test F2 landed in the same 0.60–0.62 band as V2 GBT across all 3 folds, which is exactly the consistency that told us V3 was not a meaningful improvement and did not justify rushing the blind test.

A caveat applies to Phase II: those experiments used a single 2015 quarterly split rather than the rolling year-based folds, so the Phase II numbers should be read as directional only. The Phase III CV results are the trustworthy ones.

### Key Findings

1. **Threshold tuning delivered the largest single V1 lever.** Lowering the decision boundary from 0.50 to 0.35 added +0.062 F2 to the V1 GBT (0.4974 → 0.5598). This is not a model improvement; it is an operational decision to catch more delays at the cost of more false alarms. For an airline OCC, missed delays are far more expensive than pre-allocated standby resources, so the trade-off is justified.

2. **New feature families drove the real improvement, not transformations.** In Phase II, standard transformations (standardization, binary precipitation, visibility bucketing) on 1-year data produced near-zero improvement (+0.0007). In Phase III V1, scaling to 4 years and adding PageRank, 2-4hr airport congestion, and 2-hour weather lag improved F2 over the Phase II baseline. In V2, adding weather-trend and tail-number propagation features added another +0.057 at t=0.50 and +0.071 at t=0.35. The improvements came from new predictive signal, not from re-engineering existing columns.

3. **Hyperparameter tuning was near-irrelevant.** GBT grid search across 18 configurations improved F2 by only +0.003. LR grid search produced zero improvement. LSTM grid search across 12 configurations actually under-performed the V2 GBT champion at t=0.35 (0.6118 vs 0.6336). The default model parameters were already near-optimal, confirming that the bottleneck is feature signal, not model complexity.

4. **All models converged to a narrow F2 band at t=0.50.** LR, RF, GBT, and MLP land within 0.035 F2 on the V1 blind test, and the V2 lift was directionally consistent across all four candidates. This convergence is the strongest evidence that the available feature set imposes a ceiling and that further gains require new feature families rather than new model architectures.

### V2 Result Discussion: The Biggest Gain

The V2 extended feature set produced the largest improvement in the project. At the default threshold, GBT jumped from F2 = 0.4974 (V1) to **0.6198** (V2), a +0.12 absolute F2 gain. Two new feature families drove this:

- **Weather trend features** (6h-to-2h change in temperature, wind, humidity, visibility) capture whether conditions are *worsening* — a directionally stronger predictor of delay than the static 2-hour snapshot alone.
- **Tail-number delay propagation** (`prev_DEP_DEL15`, `turnaround_minutes`) captures the cascading effect: if the same aircraft's previous leg was delayed, the next leg is more likely to be delayed, and a tight turnaround amplifies that risk.

Precision improved dramatically as well: V1 GBT delivered 30% precision at t=0.50; V2 GBT delivers **51%**. This is the most important finding of the project because it directly validated the V1 gap analysis hypothesis — missing propagation features were the precision bottleneck, and adding them moved precision by 21 absolute percentage points without sacrificing recall (recall went from 64% to 65%). The LSTM also climbed from F2 = 0.54 (V1) to 0.61 (V2), but even after grid-search hyperparameter tuning it did not overtake V2 GBT, reinforcing finding #3 above. The V3 experiment (adding absolute 2-hour weather levels back alongside trends) produced no meaningful CV improvement over V2, indicating that the trend representation already captures most of the signal in the 2-hour weather window.
